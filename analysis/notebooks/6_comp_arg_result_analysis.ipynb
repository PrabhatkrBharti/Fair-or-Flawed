## Importing dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../code/results/aggregated/overall_results.csv', index_col='Index')
```

## Visualizations

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['DF_QuAD_Score'], bins=10, kde=True, color='blue')
plt.title('Distribution of DF_QuAD_Score')
plt.xlabel('DF_QuAD_Score')

plt.subplot(1, 2, 2)
sns.histplot(df['Five_Level_Score'], bins=5, kde=True, color='green')
plt.title('Distribution of Five_Level_Score')
plt.xlabel('Five_Level_Score')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.countplot(x='Majority_Voting_Score', data=df, palette='Set2')
plt.title('Majority Voting Score Distribution')

plt.subplot(1, 2, 2)
sns.countplot(x='All_Accept_Score', data=df, palette='Set1')
plt.title('All Accept Score Distribution')

plt.tight_layout()
plt.show()
```

```python
criteria_columns = ['CLA', 'JUS', 'DEP', 'FAI', 'CON', 'ENG', 'ACC', 'CST', 'NOV', 'ETH']

criteria_counts = {}
for col in criteria_columns:
    criteria_counts[col] = df[col].value_counts()

criteria_df = pd.DataFrame(criteria_counts)
```

```python
criteria_df.T.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='Paired')
plt.title('POS, NEG, and NEU Annotations for Each Evaluation Criterion')
plt.ylabel('Count')
plt.show()
```
